```python
# import pandas and read CSV containing drug data
import pandas as pd

drug_data = pd.read_csv("PANCANCER_IC_Wed Jun 18 15_22_15 2025.csv")
drug_data
```

```python
# read CSV containing genetic data
genetic_data = pd.read_csv("PANCANCER_Genetic_features_Wed Jun 18 15_22_44 2025.csv")
genetic_data
```

# Pivot & Reset Index (Genetic Data)
Each row represents a different cell line

Each column represents a different mutation

0 = not mutated, 1 = mutated

```python
# pivot using COSMIC ID and Genetic Feature and reset index to remove extra row
genetic_data_pivot = genetic_data.pivot(index="COSMIC ID", columns="Genetic Feature", values="IS Mutated")
genetic_data_pivot_reset = genetic_data_pivot.reset_index()
genetic_data_pivot_reset
```

# Dropping NA Values
Preserved authentic data

```python
# drop NA values from pivoted genetic data
genetic_data_dropna = genetic_data_pivot_reset.dropna()
genetic_data_dropna
```

# Preparation for Joining
Targeted column for joining: COSMIC ID

ID numbers > names

Check duplicates, shared cells

```python
# rename "Cosmic ID" column in drug dataset to match case with genetic dataset
drug_data = drug_data.rename(columns={"Cosmic ID": "COSMIC ID"})
drug_data
```

```python
# check for duplicates within each dataset and make sure COSMIC IDs are shared
genetic_duplicates = genetic_data_dropna.duplicated(subset=["COSMIC ID"]).sum()
drug_duplicates = drug_data.duplicated(subset=["COSMIC ID", "Drug ID"]).sum()
shared = set(drug_data["COSMIC ID"]).intersection(set(genetic_data_dropna["COSMIC ID"]))

print("Genetic Data Duplicates: ", genetic_duplicates)
print("Drug Data Duplicates: ", drug_duplicates)
print("Shared COSMIC IDs: ", len(shared))
```

# Merging Datasets
Inner join on "COSMIC ID"

Checked original and merged shapes

```python
# merge on "COSMIC ID" column using an inner join
merged = pd.merge(drug_data, genetic_data_dropna, on="COSMIC ID", how="inner")
merged
```

```python
# check shapes of drug, genetic, and merged data to make sure they were properly joined
print(drug_data.shape)
print(genetic_data_dropna.shape)
print(merged.shape)
#print(merged.columns.tolist())
```

# Dropping Unnecessary Columns
* Drug Name dropped in favor of Drug ID
* Cell Line Name dropped in favor of COSMIC ID
* COSMIC ID no longer necessary after merge.
* TCGA Classification, Tissue, and Tissue Sub-type dropped (reduced based on available compute)
* IC50 is going to be studied. All other metrics/statistics dropped
* Dataset Version not important for study
* Remaining: Drug ID, COSMIC ID, TCGA Classification, Tissue, Tissue Sub-type, IC50, all mutations

```python
# drop unnecessary columns that will not be included in this study
X = merged.drop(["COSMIC ID", "TCGA Classification", "Tissue", "Tissue Sub-type", "Drug Name", "Cell Line Name", "AUC", "Max Conc", "RMSE", "Z score",
                             "Dataset Version"], axis=1)
X
```

# Exploring Data & Cleaning
* Basic statistics
* Dropped NA values to preserve authenticity

```python
# basic statistical analysis of merged dataset
X.describe()
```

# Checking for Outliers in IC50
Logarithmic bins

```python
# import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# isolate IC50 column, drop missing values, and filter for values greater than 0
ic50_raw = X["IC50"]
ic50 = ic50_raw.dropna()
ic50 = ic50_raw[ic50_raw > 0]

# create logarithmic bins (original data was logarithmically transformed)
bins = np.logspace(np.log10(ic50.min()), np.log10(ic50.max()), 50)

# create histogram of IC50 distribution
plt.figure(figsize=(12,6))
sns.histplot(ic50, bins=bins, kde=False)
plt.title("Distribution of IC50")
plt.xscale("log")
plt.xlabel("IC50 (Micromolar)")
plt.ylabel("Frequency")
plt.savefig("Distribution of IC50.png")
plt.show()
```

# Encoding
OneHotEncoding of categorical column (Drug ID)

```python
# import OHE library
from sklearn.preprocessing import OneHotEncoder

# isolate single categorical column
categorical_columns = ["Drug ID"]

# create OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")

# fit and transform categorical columnss to OneHotEncoder and retain Drug IDs
X_enc = ohe.fit_transform(X[categorical_columns])
enc_names = ohe.get_feature_names_out(categorical_columns)
X_categorical = pd.DataFrame(X_enc, columns=enc_names, index=X.index)
X_categorical
```

# Numeric Columns (ICI50)

```python
# combine pivoted mutation columns and IC50 column (numeric columns)
mutation_columns = [col for col in X.columns if col.endswith('_mut')]
numeric_columns = ["IC50"] + mutation_columns
X_numeric = X[numeric_columns]
X_numeric
```

```python
# combine categorical (OneHotEncoded) and numerical columns together
X_combined = pd.concat([X_categorical, X_numeric], axis=1)
X_combined
```

# Splitting into Features & Target

```python
# split into features (drugs and mutations) and target (IC50)
x = X_combined.drop(columns=["IC50"])
y = X_combined["IC50"]
```

# Splitting into Testing and Training Sets
80% training, 20% testing

```python
# import library and create testing and training sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

# Creating a Function to Evaluate Models

```python
# import libraries for statistical analysis
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error

# create a function to quickly determine r-squared, mean squared error, and root mean squared error for model evaluation
def evaluate_model(model, x, y):
    preds = model.predict(x)
    print("R-Squared: ", r2_score(y, preds))
    print("MSE: ", mean_squared_error(y, preds))
    print("RMSE: ", root_mean_squared_error(y, preds))
```

# Random Forest Regression

```python
# import library
from sklearn.ensemble import RandomForestRegressor

# create random forest model with initial parameters
rfr_model = RandomForestRegressor(
    random_state=42,
    n_estimators=100,
    max_depth=None,
    n_jobs=-1
)

# fit training data to random forest model
rfr_model.fit(X_train, y_train)

# evaluate model
evaluate_model(rfr_model, X_test, y_test)

# play sound when cell has run
Audio(sound_file, autoplay=True)
```

# Visualizing Random Forest Regression

```python
# import libraries
import matplotlib.pyplot as plt

# predict target using random forest model with X_test
y_pred_rfr = rfr_model.predict(X_test)

# create scatterplot of actual vs. predicted values
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred_rfr, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "--r")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs. Actual Values in Random Forest Regression")
plt.savefig("Predicted vs. Actual Values in Random Forest Regression.png")
```

```python
# calculate residuals for predictions
rfr_residuals = y_test - y_pred_rfr

# create residual plot
plt.figure(figsize=(6, 6))
plt.scatter(y_pred_rfr, rfr_residuals, alpha=0.5)
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot for Random Forest Regression")
plt.savefig("Residual Plot for Random Forest Regression.png")
plt.show()
```

# Scaling
Necessary for PCA, Ridge Regression, and Lasso Regression

```python
# import libraries
from sklearn.preprocessing import StandardScaler

# create scaler and scale training and testing sets
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)
```

# PCA
Goal: Reduce extreme number of features

```python
# import libraries

from sklearn.decomposition import PCA

# create PCA while preserving 50% explained variance
pca = PCA(n_components=0.50)

# fit and transform to scaled training data
X_train_pca = pca.fit_transform(X_train_scaled)

# compare original and PCA'd features and calculate actual explained variance
print("Original Features: ", X_train.shape[1])
print("Reduced Features: ", X_train_pca.shape[1])
print("Explained Variance: ", sum(pca.explained_variance_ratio_))
```

# Tuning Random Forest Regression

```python
# import libraries
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# set parameters for RandomizedSearchCV to use
params = {
    "n_estimators": randint(100, 200),
    "max_depth": [None, 15, 30],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features" : ["sqrt"]
}

# create search
search = RandomizedSearchCV(
    random_state=42,
    estimator = rfr_model,
    param_distributions = params,
    n_iter = 30,
    scoring = "r2",
    cv = 3,
    verbose = 2,
    n_jobs = -1
)

# fit search to scaled and PCA'd data and print best parameters for use in tuned model
search.fit(X_train_pca, y_train)
print(search.best_params_)

# play sound when cell has run
Audio(sound_file, autoplay=True)
```

```python
# create tuned model with best parameters from RandomizedSearchCV
rfr_model_tuned = RandomForestRegressor(
    random_state=42,
    n_estimators=146,
    max_depth=None,
    min_samples_leaf=2,
    min_samples_split=5,
    max_features="sqrt",
    n_jobs=-1
)

# fit tuned model to data
rfr_model_tuned.fit(X_train, y_train)

# evaluate model performance
evaluate_model(rfr_model_tuned, X_test, y_test)

# play audio when cell has run
Audio(sound_file, autoplay=True)
```

# Visualizing Tuned Random Forest Regression

```python
# predict target using tuned random forest model model with X_test and create scatterplot

y_pred_rfr_tuned = rfr_model_tuned.predict(X_test)

plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred_rfr_tuned, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "--r")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs. Actual Values in Tuned Random Forest Regression")
plt.savefig("Predicted vs. Actual Values in Tuned Random Forest Regression.png")
```

```python
# calculate residuals and create residual plot for tuned random forest regression model

rfr_tuned_residuals = y_test - y_pred_rfr_tuned

plt.figure(figsize=(6, 6))
plt.scatter(y_pred_rfr_tuned, rfr_tuned_residuals, alpha=0.5)
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot for Tuned Random Forest Regression")
plt.savefig("Residual Plot for Tuned Random Forest Regression.png")
plt.show()
```

# Neural Network Regression

```python
# import library
from sklearn.neural_network import MLPRegressor

# create first neural network with base parameters and ReLu activation
nn1 = MLPRegressor(
    random_state=42,
    activation="relu",
    hidden_layer_sizes=(10, 100),
    alpha=0.001,
    early_stopping=True,
    max_iter=1000
)

# fit neural network to training data
nn1.fit(X_train_scaled, y_train)

# evalute initial neural network
evaluate_model(nn1, X_test_scaled, y_test)

# play sound when complete
Audio(sound_file, autoplay=True)
```

# Visualizing Neural Network Regression

```python
# predict target using neural network with X_test and create scatterplot

y_pred_nn1 = nn1.predict(X_test_scaled.values)

plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred_nn1, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "--r")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs. Actual Values in Neural Network")
plt.savefig("Predicted vs. Actual Values in Neural Network.png")
```

# Tuning Neural Network Regression

```python
# import library
from sklearn.model_selection import RandomizedSearchCV

# create new neural network with early stopping and a maximum of 1000 iterations
nn2 = MLPRegressor(
    early_stopping=True,
    max_iter=1000
)

# set parameters for RandomizedSearchCV to use
nn_params = {
    "activation": ["relu", "tanh"],
    "hidden_layer_sizes": [(50,), (100,), (50,50), (100, 50)],
    "alpha": [0.0001, 0.001, 0.01, 0.1],
    "learning_rate_init": [0.001, 0.01, 0.1],
#    "solver": ["adam", "lbfgs"]
}

# begin search using parameters scored by r-squared
nn_search = RandomizedSearchCV(
    random_state=42,
    estimator = nn2,
    param_distributions = nn_params,
    n_iter = 20,
    scoring = "r2",
    cv = 3,
    verbose = 2,
    n_jobs = 10
)

# fit search to training data
nn_search.fit(X_train_scaled, y_train)

# evalute model with best parameters
nn_tuned = nn_search.best_estimator_
evaluate_model(nn_tuned, X_test_scaled, y_test)

# play sound when complete
Audio(sound_file, autoplay=True)
```

# Visualizing Tuned Neural Network Regression

```python
# predict target using tuned neural network with X_test and create scatterplot

y_pred_nn_tuned = nn_tuned.predict(X_test_scaled.values)

plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred_nn_tuned, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "--r")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs. Actual Values in Tuned Neural Network")
plt.savefig("Predicted vs. Actual Values in Tuned Neural Network.png")
```

```python
# calculate residuals and create residual plot for tuned neural network model

nn_tuned_residuals = y_test - y_pred_nn_tuned

plt.figure(figsize=(6, 6))
plt.scatter(y_pred_nn_tuned, nn_tuned_residuals, alpha=0.5)
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot for Tuned Neural Network")
plt.savefig("Residual Plot for Tuned Neural Network.png")
plt.show()
```

# Simpler Model for Deployment (CatBoost)

```python
# import library and datasets in the event prior cells were not run

import pandas as pd

drug_data = pd.read_csv("PANCANCER_IC_Wed Jun 18 15_22_15 2025.csv")
genetic_data = pd.read_csv("PANCANCER_Genetic_features_Wed Jun 18 15_22_44 2025.csv")
```

```python
# filter columns from drug data
drug_data_cat = drug_data[["Drug Name", "Cosmic ID", "IC50"]]

# match case with genetic data for column used with join
drug_data_cat = drug_data_cat.rename(columns={"Cosmic ID": "COSMIC ID"})
```

```python
# filter columns from genetic data
genetic_data_cat = genetic_data[["COSMIC ID", "Genetic Feature"]]
```

```python
# merge drug and genetic data on "COSMIC ID" using an inner join
merged_cat = pd.merge(drug_data_cat, genetic_data_cat,
                        on="COSMIC ID", how="inner")
```

```python
# remove "COSMIC ID" and reorder necessary columns
merged_cat_selected = merged_cat[["Drug Name", "Genetic Feature", "IC50"]]

# save merged dataset to CSV for deployment
# merged_cat_selected.to_csv("merged_cat_selected.csv")
```

```python
# check for missing values
merged_cat_selected.isna().sum()
```

```python
# reduce dataset size based on available compute using a random sample
merged_cat_sample = merged_cat_selected.sample(n=200000)
```

```python
# identify features and target variables

features_cat = merged_cat_sample[["Drug Name", "Genetic Feature"]]
target_cat = merged_cat_sample[["IC50"]]
```

```python
# import libray and create testing and training datasets

from sklearn.model_selection import train_test_split

X_train_cat, X_test_cat, y_train_cat, y_test_cat = train_test_split(features_cat, target_cat, test_size=0.2)
```

```python
# import library
from catboost import CatBoostRegressor

# identify categorical columns for CatBoostRegressor
cat_columns = ["Drug Name", "Genetic Feature"]

# create CatBoost Regressor model
cat_model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.01,
    depth=6,
    task_type="GPU",
    devices="0",
    verbose=100
)

# fit model to shortened training data and identify categorical columns
cat_model.fit(X_train_cat, y_train_cat, cat_features=cat_columns)

# evaluate model performance
evaluate_model(cat_model, X_test_cat, y_test_cat)

# play sound when complete
Audio(sound_file, autoplay=True)
```

```python
# predict target using CatBoost model with X_test and create scatterplot

import matplotlib.pyplot as plt

y_pred_cat = cat_model.predict(X_test_cat)

plt.figure(figsize=(6, 6))
plt.scatter(y_test_cat, y_pred_cat, alpha=0.5)
plt.plot([y_test_cat.min(), y_test_cat.max()], [y_test_cat.min(), y_test_cat.max()], "--r")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs. Actual Values in CatBoost Regression")
plt.savefig("Predicted vs. Actual Values in Tuned CatBoost Regression.png")
```

```python
# save model for deployment
cat_model.save_model("catboost_model.cbm")
```
